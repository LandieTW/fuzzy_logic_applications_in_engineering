Two types of steel are tested four times each for their tensile strength. Let us consider fuzzy sets 'A' and 'B' to be the two types of steel on the universe of strengths (no dimensional units) where four tests X = {1, 2, 3, 4} were conducted on each of the steel types. The following given sets represent the tensile strength of each steel. Compare the tensile strength of these two steels by finding the following.
A = {0.4/1 + 0.35/2 + 0.5/3 + 0.6/4}
B = {0.7/1 + 0.75/2 + 0.65/3 + 0.8/4}
a. A intersection B
b. A union B
c. complement A and complement B
d. A difference B

```python
a = [.4, .35, .5, .6]
b = [.7, .75, .65, .8]
days = [1, 2, 3, 4]
```

```python
a_intersection_b = [f"{min(a_i, b_i)}/{day}" for a_i, b_i, day in zip(a, b, days)]
print(a_intersection_b)
```

```python
a_union_b = [f"{max(a_i, b_i)}/{day}" for a_i, b_i, day in zip(a, b, days)]
print(a_union_b)
```

```python
complement_a = [f"{1-a_i}/{day}" for a_i, day in zip(a, days)]
print(complement_a)
```

```python
complement_b = [f"{1-b_i}/{day}" for b_i, day in zip(b, days)]
print(complement_b)
```

```python
a_difference_b = [f"{min(a_i, 1-b_i)}/{day}" for a_i, b_i, day in zip(a, b, days)]
print(a_difference_b)
```
