If we have an ellipse, with parameters 'a' and 'b', under what conditions of these two parameters will a general elliptic shape became a circle?

Mathematically, we know that...
when a/b = 1, a circle results 
when a/b << 1 or a/b >> 1, we have an elliptic shape
a/b -> infinity, a line segment results
