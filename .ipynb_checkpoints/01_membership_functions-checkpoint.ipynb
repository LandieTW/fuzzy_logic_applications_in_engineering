Developing a reasonable membership function for the following fuzzy sets based on moving vehicles on a freeway for the speed range of 0-75 mph.
a. Fast
b. Moderate
c. Slow

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

```python
speed = ctrl.Antecedent(np.arange(0, 75, 1), 'speed')
```

```python
speed.automf(names=['fast', 'moderate', 'slow'])
```

```python
speed['fast'] = fuzz.trapmf(speed.universe, 
                            [37.5, 50, 62.5, 75])
speed['moderate'] = fuzz.trimf(speed.universe,
                               [25, 37.5, 50])
speed['slow'] = fuzz.trapmf(speed.universe,
                            [0, 12.5, 25, 37.5])
```

```python
%matplotlib inline

speed.view()
plt.title("Speed membership function")
plt.show()
```
