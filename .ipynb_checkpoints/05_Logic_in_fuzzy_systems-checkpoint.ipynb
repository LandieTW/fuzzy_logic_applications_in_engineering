Consider the following two discrete fuzzy sets, which are defined on a universe X = [0, 9]
A = {.4/0 + .7/1 + .2/2 + .1/3 + .3/4}
B = {0/5 + .5/6 + .6/7 + .3/8 + .7/9}
b. Construct the relation for the rule: 'IF A, THEN B' using Mandani implication
c. If we introduce a new antecedent
A'= {.3/0 + .7/1 + .6/2 + .3/3 + .5/4}
find the new consequent B', using max-min composition

```python
import pandas as pd
```

```python
a_universe = range(5)
a_values = [.4, .7, .2, .1, .3]
b_universe = range(5, 10)
b_values = [.0, .5, .6, .3, .7]
```

```python
# mandani = min[a-x, b-y]

a_b = [
    [
        min([a_values[i], b_values[j]])
        for j in range(len(a_universe))
    ]
    for i in range(len(b_universe))
]

mandani = pd.DataFrame(
    data=a_b,
)

print(mandani)
```

```python
a2_universe = range(5)
a2_values = [.3, .7, .6, .3, .5]
```

```python
# b2_values = a2 . mandani

min_a2_mandani = [
    [
        min(a2_values[i], mandani[i][j]) 
        for j in range(len(a2_universe))
    ] for i in range(len(a2_universe))
]

max_a2_mandani = [
    max(min_a2_mandani[k])
    for k in range(len(min_a2_mandani))
]

b2_values = [
    f"{max_a2_mandani[h]}/{b_universe[h]}"
    for h in range(len(b_universe))
]

print(b2_values)
```
