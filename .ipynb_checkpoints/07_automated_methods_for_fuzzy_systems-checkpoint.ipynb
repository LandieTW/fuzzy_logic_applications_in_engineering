```python
X_train = np.column_stack([major_train, minor_train])
y_train = sigma_train

X_test = np.column_stack([major_test, minor_test])
y_test = sigma_test
```

```python
def fuzzy_bls_algo(
    X_train: np.column_stack,
    y_train: np.array,
    X_test: np.column_stack,
    y_test: np.array,
    n_sets: int = 3
) -> np.array:
    """
    Calculates fuzzy model with BLS algorithm

    Args:
        X_train: Train input
        y_train: Train output
        X_test: Test input
        y_test: Test output

    Returns:
        b: consequent parameter
        train_pred: train prediction
        test_pred: test prediction
    """

    # Normalizing data
    x1_train_norm = (
        (X_train[:, 0] - np.mean(X_train[:, 0])) / np.std(X_train[:, 0])
    )
    x2_train_norm = (
        (X_train[:, 1] - np.mean(X_train[:, 1])) / np.std(X_train[:, 1])
    )
    x1_test_norm = (
        (X_test[:, 0] - np.mean(X_test[:, 0])) / np.std(X_test[:, 0])
    )
    x2_test_norm = (
        (X_test[:, 1] - np.mean(X_test[:, 1])) / np.std(X_test[:, 1])
    )

    # number of rules
    R = n_sets ** 2

    # defining the centers
    c1 = np.linspace(np.min(x1_train_norm), np.max(x1_train_norm), n_sets)
    c2 = np.linspace(np.min(x2_train_norm), np.max(x2_train_norm), n_sets)

    # Defining widths
    sigma1 = (np.max(x1_train_norm) - np.min(x1_train_norm)) / n_sets
    sigma2 = (np.max(x2_train_norm) - np.min(x2_train_norm)) / n_sets

    # Matrix
    def matrix(
        x1_norm: np.column_stack,
        x2_norm: np.column_stack
    ) -> np.array:
        """
        Generate the regression matrix for fuzzy model

        Args:
            x1_norm: input 1 normalised
            x2_norm: input 2 normalised

        Returns:
            Phi matrix
        """
        N = len(x1_norm)
        Phi = np.zeros((N, R))

        for i in range(N):
            w = list()
            for j in range(n_sets):
                for k in range(n_sets):

                    # membership functions
                    mu_1 = np.exp(-.5 * (x1_norm[i] - c1[j]) / sigma1)**2
                    mu_2 = np.exp(-.5 * (x2_norm[i] - c2[j]) / sigma2)**2

                    # rule weight
                    w.append(mu_1 * mu_2)
                    
            # sum the weights
            w_sum = sum(w)
            
            if w_sum == 0:
                w_sum = 1e-10   # avoid zero division
                
            Phi[i, :] = [w_i / w_sum for w_i in w]

        return Phi

    Phi_train = matrix(x1_train_norm, x2_train_norm)

    # bls = (ΦᵀΦ)⁻¹ Φᵀ y
    bls = np.linalg.pinv(Phi_train.T @ Phi_train) @ Phi_train.T @ y_train

    y_train_pred = Phi_train @ bls

    Phi_test = matrix(x1_test_norm, x2_test_norm)
    y_test_pred = Phi_test @ bls

    return bls, y_train_pred, y_test_pred, c1, c2, sigma1, sigma2
```

```python
bls, y_train_pred, y_test_pred, c1, c2, sigma1, sigma2 = fuzzy_bls_algo(
    X_train, y_train, X_test, y_test, n_sets=3
)
```

```python
# Errors

rmse_train = np.sqrt(np.mean((y_train - y_train_pred)**2))
rmse_test = np.sqrt(np.mean((y_test - y_test_pred)**2))

mae_train = np.mean(np.abs(y_train - y_train_pred))
mae_test = np.mean(np.abs(y_test - y_test_pred))
```

RECURSIVE LEAST SQUARES ALGORITHM

RLS ALGO

The RLS algorithm is similar to the BLS algorithm; however, the RLS algorithm makes updating theta much easier. The algorithm is a recursive version of the BLS method. It operates without using all the training data and most importantly without having to compute the inverse of ΦᵀΦ each time the theta is updated. RLS calculates theta(k) at each time step k from the past estimate theta(k-1) and the latest data pair that is received, (x, yk).

```python
def fuzzy_rls_algo(
    X_train: np.column_stack,
    y_train: np.array,
    X_test: np.column_stack,
    y_test: np.array,
    n_sets: int = 3,
    lambda_: float = .99,
    delta: float = 1_000.0
) -> np.array:
    """
    Calculates fuzzy model with RLS algorithm

    Args:
        X_train: Train input
        y_train: Train output
        X_test: Test input
        y_test: Test output
        n_sets: number of fuzzy sets for each variable
        lambda_: forget factor
        delta: initial value for P matrix

    Returns:
        theta: estimated parameters
        train_pred: train prediction
        test_pred: test prediction
        theta_history: parameters history
    """

    # Norm data
    x1_mean = np.mean(X_train[:, 0])
    x1_std = np.std(X_train[:, 0])
    x2_mean = np.mean(X_train[:, 1])
    x2_std = np.std(X_train[:, 1])

    x1_train_norm = (X_train[:, 0] - x1_mean) / x1_std
    x2_train_norm = (X_train[:, 1] - x2_mean) / x2_std

    x1_test_norm = (X_test[:, 0] - x1_mean) / x1_std
    x2_test_norm = (X_test[:, 1] - x2_mean) / x2_std

    # Number of rules
    R = n_sets ** 2

    # Defining centers
    c1 = np.linspace(np.min(x1_train_norm), np.max(x1_train_norm), n_sets)
    c2 = np.linspace(np.min(x2_train_norm), np.max(x2_train_norm), n_sets)

    # Defining widths
    sigma1 = (np.max(x1_train_norm) - np.min(x1_train_norm)) / n_sets
    sigma2 = (np.max(x2_train_norm) - np.min(x2_train_norm)) / n_sets

    def create_phi_vector(
        x1_norm: np.array,
        x2_norm: np.array,
        c1: np.linspace,
        c2: np.linspace,
        sigma1: np.array,
        sigma2: np.array,
        n_sets: int = 3
    ) -> np.array:
        """
        Creates phi vector (regressor) for each point

        Args:
            x1_norm: input 1 normalised
            x2_norm: input 2 normalised
            c1: centers 1 
            c2: centers 2
            sigma1: center's 1 width
            sigma2: center's 2 width
            n_sets: number of fuzzy sets for each variable
        """
        R = n_sets ** 2
        phi = np.zeros(R)
        w = list()
    
        for i in range(n_sets):
            for j in range(n_sets):

                # membership functions
                mu1 = np.exp(-.5 * ((x1_norm - c1[i]) / sigma1) ** 2)
                mu2 = np.exp(-.5 * ((x2_norm - c2[j]) / sigma2) ** 2)
                w.append(mu1 * mu2)

        w_sum = np.sum(w)
        if w_sum == 0:
            w_sum = 1e-10

        phi = np.array([w_i / w_sum for w_i in w])
        return phi

    # initialization of RLS parameters
    theta = np.zeros(R)
    P = delta * np.eye(R)
    lambda_inv = 1.0 / lambda_

    # analysis history
    theta_history = np.zeros((len(y_train), R))
    y_train_pred = np.zeros(len(y_train))
    
    # RLS Training (Sequencial data)
    for i in range(len(y_train)):

        # creating vector
        phi_i = create_phi_vector(
            x1_train_norm[i], x2_train_norm[i],
            c1, c2, sigma1, sigma2, n_sets
        )

        y_pred_i = np.dot(phi_i, theta)
        error = y_train[i] - y_pred_i

        # Kalman Gain
        phi_P = np.dot(P, phi_i)
        denom = lambda_ + np.dot(phi_i, phi_P)
        K = phi_P / denom

        # Updating parameters
        theta = theta + K * error

        # Updating matrix P
        P = (P - np.outer(K, phi_P)) * lambda_inv

        # Storage hystory
        theta_history[i, :] = theta
        y_train_pred[i] = y_pred_i

        # Showing progress
        if i % 5 == 0:
            print(f"Sample {i+1}/{len(y_train)}: MSE = {error**2:.6f}")

    # Training data set predictions
    y_train_pred_final = np.zeros(len(y_train))
    for i in range(len(y_train)):
        phi_i = create_phi_vector(
            x1_train_norm[i], x2_train_norm[i],
            c1, c2, sigma1, sigma2, n_sets
        )
        y_train_pred_final[i] = np.dot(phi_i, theta)

    # Testing data set predictions
    y_test_pred = np.zeros(len(y_test))
    for i in range(len(y_test)):
        phi_i = create_phi_vector(
            x1_test_norm[i], x2_test_norm[i],
            c1, c2, sigma1, sigma2, n_sets
        )
        y_test_pred[i] = np.dot(phi_i, theta)

    return theta, y_train_pred_final, y_test_pred, theta_history, c1, c2, sigma1, sigma2
```

```python
# Execute RLS modeling

lambda_factor = .99

theta_rls, y_train_pred_rls, y_test_pred_rls, theta_history, c1, c2, sigma1, sigma2 = fuzzy_rls_algo(
    X_train, y_train, X_test, y_test, n_sets=3, lambda_=lambda_factor, delta=1_000.0
)
```

```python
# Errors

rmse_train = np.sqrt(np.mean((y_train - y_train_pred_rls) ** 2))
rmse_test = np.sqrt(np.mean((y_test - y_test_pred_rls) ** 2))

mae_train = np.mean(np.abs(y_train - y_train_pred_rls))
mae_test = np.mean(np.abs(y_test - y_test_pred_rls))
```

GRADIENT METHOD

In the RLS method, the predicted output for the training data set could have been improved and recommended modifying the input parameters. The GM dos just that and provides a means for tuning the parameters of the fuzzy model (the parameters of the rule-base). Recall that for the input membership function, we have the membership function centers and the spread of the membership functions. In addition to the input parameters, the GM provides a method to tune the output membership function.

```python
def fuzzy_gradient_method(
    X_train: np.column_stack,
    y_train: np.array,
    X_test: np.column_stack,
    y_test: np.array,
    n_sets: int = 3,
    learning_rate_b: float = .01,
    learning_rate_c: float = .01,
    learning_rate_sigma: float = .001,
    epochs: int = 1_000,
    tol: float = 1e-6,
) -> np.array:
    """
    Calculates the gradient method for fuzzy models

    1. Normalization of the input data
    2. Initialization of the fuzzy model parameters
    3. Use a descendent gradient to adjust the parameters
    4. Minimization of MSQE between predictions and real values
    5. Return the model and the predictions

    Args:
    -----
    X_train: array (N, 2)
        Train input: Major Stress | Minor Stress
    y_train: array (N, )
        Train output: SigmaL 
    X_test: array (M, 2)
        Test input: Major Stress | Minor Stress
    y_test: array (M, )
        Test output: SigmaL
    n_sets: 
        number of fuzzy sets for each variable
    learning_rate_b:
        learning rate for consequence parameter
    learning_rate_c:
        learning rate for centers parameters
    learning_rate_sigma:
        learning rate for widths parameters
    epochs: 
        Maximum number of iteractions of the gradient
    tol:
        Stop criteria (Convergence)
    
    Returns
    -------
    b: array (n_sets * n_sets)
        Optimized consequent parameter
    c1, c2: array (n_sets)
        Optimized centers for both variables
    sigma1, sigma2: array (n_sets)
        Optimized widths for both variables
    y_train_pred: array (N, )
        Model predictions for the training data
    y_test_pred: array (M, )
        Model predictions for the test data
    losses:
        Error hystory in each training step
    """

    # Normalization of the inputs
    x1_mean = np.mean(X_train[:, 0])
    x1_std = np.std(X_train[:, 0])
    x2_mean = np.mean(X_train[:, 1])
    x2_std = np.std(X_train[:, 1])
    
    x1_train_norm = (X_train[:, 0] - x1_mean) / x1_std
    x2_train_norm = (X_train[:, 1] - x2_mean) / x2_std

    x1_test_norm = (X_test[:, 0] - x1_mean) / x1_std
    x2_test_norm = (X_test[:, 1] - x2_mean) / x2_std

    # Initialization of the fuzzy parameters
    R = n_sets ** 2

    c1 = np.linspace(np.min(x1_train_norm), np.max(x1_train_norm), n_sets)
    c2 = np.linspace(np.min(x2_train_norm), np.max(x2_train_norm), n_sets)

    sigma1 = np.ones(n_sets) * (np.max(x1_train_norm) - np.min(x1_train_norm)) / n_sets
    sigma2 = np.ones(n_sets) * (np.max(x2_train_norm) - np.min(x2_train_norm)) / n_sets

    b = np.zeros(R)

    losses = list()

    # Main loop of the gradient
    for epoch in range(epochs):
        
        loss = 0
        grad_b = np.zeros(R)
        grad_c1 = np.zeros(n_sets)
        grad_c2 = np.zeros(n_sets)
        grad_sigma1 = np.zeros(n_sets)
        grad_sigma2 = np.zeros(n_sets)

        # loop above each training point (Batch Gradient)
        for i in range(len(x1_train_norm)):

            # for each fuzzy-set combination, 
            # calculates how much the rule applyes to the actual point
            w = list()
            mu1_values = list()    # membership functions for x1
            mu2_values = list()    # membership functions for x2

            # loob abot all fuzzy set combinations
            for j in range(n_sets):
                for k in range(n_sets):
                    mu1 = np.exp(-.5 * ((x1_train_norm[i] - c1[j]) / sigma1[j]) ** 2)
                    mu2 = np.exp(-.5 * ((x2_train_norm[i] - c2[k]) / sigma2[k]) ** 2)
                    mu1_values.append(mu1)
                    mu2_values.append(mu2)
                    w.append(mu1 * mu2)

            # Normalization of the sum
            w_sum = np.sum(w)
            if w_sum == 0:
                w_sum = 1e-10   # Avoid zero division

            # Vetor phi - normalized weigths
            phi = np.array([w_i / w_sum for w_i in w])

            # predictions with actual parameters
            y_pred = np.dot(phi, b)
            error = y_pred - y_train[i]
            loss += error ** 2

            # Calculating the gradient (chain rule)
            common_factor = 2 * error
            for idx in range(R):
                j = idx // n_sets
                k = idx % n_sets

                phi_idx = phi[idx]
                mu1 = mu1_values[idx]
                mu2 = mu2_values[idx]

                # gradient for consequen
                grad_b[idx] += common_factor * phi_idx

                common_term = common_factor * (b[idx] - y_pred) * (phi_idx / w_sum)
                if sigma1[j] != 0:
                    grad_c1[j] += common_term * mu2 * mu1 * (x1_train_norm[i] - c1[j]) / (sigma1[j] ** 2)
                    grad_sigma1[j] += common_term * mu2 * mu1 * ((x1_train_norm[i] - c1[j]) ** 2) / (sigma1[j] ** 3)
                if sigma2[k] != 0:
                    grad_c2[k] += common_term * mu2 * mu1 * (x2_train_norm[i] - c2[k]) / (sigma2[k] ** 2)
                    grad_sigma2[k] += common_term * mu2 * mu1 * ((x2_train_norm[i] - c2[k]) ** 2) / (sigma2[k] ** 3)

        loss /= len(y_train)
        losses.append(loss)

        grad_b /= len(y_train)
        grad_c1 /= len(y_train)
        grad_c2 /= len(y_train)
        grad_sigma1 /= len(y_train)
        grad_sigma2 /= len(y_train)

        # Updating parameters
        # new_parameter = old_parameter - learn_rate * gradient

        b_old = b.copy()
        c1_old = c1.copy()
        c2_old = c2.copy()

        b -= learning_rate_b * grad_b
        c1 -= learning_rate_c * grad_c1
        c2 -= learning_rate_c * grad_c2
        sigma1 -= learning_rate_sigma * grad_sigma1
        sigma2 -= learning_rate_sigma * grad_sigma2

        # Verifying convergence
        change_b = np.linalg.norm(b - b_old)
        change_c1 = np.linalg.norm(c1 - c1_old)
        change_c2 = np.linalg.norm(c2 - c2_old)
        total_change = change_b + change_c1 + change_c2

        if total_change < tol and epoch > 100:
            print("Converged")
            break

        if (epoch + 1) % (epochs / 10) == 0:
            print(f"Epoch: {epoch+1}/{epochs}: Loss = {loss:.8f}")
            
    def predict(
        X_norm1: np.column_stack, 
        X_norm2: np.column_stack
    ) -> np.array:
        """
        Make predictions based on the fuzzy model
    
        Args
        -----
        X_norm1
            Major stress values normalised
        X_norm2
            Minor stress values normalised
    
        Returns
        --------
            Predictions for SigmaL for each input
        """
        predictions = list()
    
        for i in range(len(X_norm1)):
            w = list()
            for j in range(n_sets):
                for k in range(n_sets):
                    mu1 = np.exp(-.5 * ((X_norm1[i] - c1[j]) / sigma1[j]) ** 2)
                    mu2 = np.exp(-.5 * ((X_norm2[i] - c2[k]) / sigma2[k]) ** 2)
                    w.append(mu1 * mu2)
        
            w_sum = np.sum(w)
            if w_sum == 0:
                w_sum = 1e-10
    
            phi = np.array([w_i / w_sum for w_i in w])
    
            prediction = np.dot(phi, b)
            predictions.append(prediction)
    
        return np.array(predictions)

    # final predictions
    y_train_pred_norm = predict(x1_train_norm, x2_train_norm)
    y_test_pred_norm = predict(x1_test_norm, x2_test_norm)

    # Removing normalization

    y_train_std = np.std(y_train)
    y_test_std = np.std(y_test)
    y_train_mean = np.mean(y_train)
    y_test_mean = np.mean(y_test)

    y_train_pred = y_train_pred_norm * y_train_std + y_train_mean
    y_test_pred = y_test_pred_norm * y_test_std + y_test_mean

    return b, c1, c2, sigma1, sigma2, y_train_pred, y_test_pred, losses
```

```python
b, c1, c2, sigma1, sigma2, y_train_pred, y_test_pred, losses = fuzzy_gradient_method(
    X_train, y_train, X_test, y_test, n_sets=3,
    learning_rate_b=.1, learning_rate_c=.005, learning_rate_sigma=.001, epochs=6_000
)
rmse_train = np.sqrt(np.mean((y_train - y_train_pred) ** 2))
rmse_test = np.sqrt(np.mean((y_test - y_test_pred) ** 2))
```

```python
print("Epoch number: ", len(losses))
print(f"Final loss: {losses[-1]:.8f}")

print("\nOptimized parameters")
print(f"C1, C2: {c1} ; {c2}")
print(f"Sigma1, Sigma2: {sigma1} ; {sigma2}")

print("\nConsequent parameters")
for i in range(len(b)):
    print(f"b{i+1}: {b[i]:.6f}")

print(f"\nErrors")
print("Training RMSE: {:.6f} ; Testing RMSE: {:.6f}".format(rmse_train, rmse_test))
```

CLUSTERING METHOD

Fuzzy clustering is the partitioning of a collection of data into fuzzy subsets or clusters based on similarities between the data. The CM, like the other methods, develops a fuzzy estimation model to predict the output given the input. The algorithm forms rules (or clusters) with training data using a nearest neighbor approach for the fuzzy system.

```python
from sklearn.metrics import pairwise_distances
```

```python
def fuzzy_subtractive_clustering_method(
    X_train: np.column_stack,
    y_train: np.array,
    X_test: np.column_stack,
    y_test: np.array,
    ra: float = .3,
    rb: float = 1.5,
    epsilon: float = 1e-4,
    max_clusters: int = 10
) -> np.array:
    """
    Implements the fuzzy model with subtractive clustering

    1. Each point is considered a potential cluster center
    2. Calculates the density for each point
    3. The high density point becomes the first center
    4. Remove density from near points
    5. Repeat it until every point has its density below some limit
    
    Args
    ------
    X_train: array (N, 2)
        Train input
    y_train: array (N, )
        Train output
    X_test: array (M, 2)
        Test input
    y_test: array (M, )
        Test output
    ra: minor values -> more clusters
        Cluster's influency radius
    rb:
        Density subtract radius
    epsilon
        Stop limit
    max_clusters
        Maximum number of clusters allowed
    
    Returns
    ----------
    clusters_centers: array (n_clusters, 2)
        Clusters centers found in the input space
    sigma_clusters: array (n_clusters, 2)
        Widths / Deviations of clusters in each dimension
    y_train_pred: array (N, )
        Training predictions
    y_test_pred: array (M, )
        Testing predictions
    n_clusters
        Number of clusters
    U: array (N, n_clusters)
        Membership matrix of the clusters
    """

    x_mean = np.mean(X_train, axis=0)
    x_std = np.std(X_train, axis=0)
    X_train_norm = (X_train - x_mean) / x_std
    X_test_norm = (X_test - x_mean) / x_std

    y_mean = np.mean(y_train)
    y_std = np.std(y_train)
    y_train_norm = (y_train - y_mean) / y_std

    def subtractive_clustering(
        X: np.array,
        ra: float,
        rb: float,
        epsilon: float,
        max_clusters: int
    ) -> np.array:
        """
        Implements the subtractive clustering algorithm
        """
        n_samples = X.shape[0]

        # calc. distance between points
        dist_matrix = pairwise_distances(X, metric='sqeuclidean')

        # calc. initial density for each points
        density = np.zeros(n_samples)
        for i in range(n_samples):
            influence = np.exp(-dist_matrix[i, :] / (ra / 2) ** 2)
            density[i] = np.sum(influence)

        # 
        cluster_centers = list()
        cluster_index = list()
        density_copy = density.copy()

        # stop criteria
        max_density_init = np.max(density_copy)
        stop_threshold = epsilon * max_density_init

        # finding max density point
        while len(cluster_centers) < max_clusters:
            idx_center = np.argmax(density_copy)
            max_density = density_copy[idx_center]
            if max_density < stop_threshold:
                print(f"Max density {max_density:.4f} < threshold: {stop_threshold:.4f}")
                break

            # adding as the cluster center
            cluster_centers.append(X[idx_center])
            cluster_index.append(idx_center)

            # Reducing density from near points
            for i in range(n_samples):
                dist_to_center = dist_matrix[idx_center, i]
                subtract_amount = max_density * np.exp(-dist_to_center / (rb / 2) ** 2)
                density_copy[i] -= subtract_amount

            # The selected point won't be chosen anymore
            density_copy[idx_center] = -np.inf

            print(f"Cluster: {len(cluster_centers)} - point: {idx_center} - density: {max_density:.4f}")

        return np.array(cluster_centers), np.array(cluster_index)

    cluster_centers, cluster_index = subtractive_clustering(
        X_train_norm, ra, rb, epsilon, max_clusters
    )

    n_clusters = len(cluster_centers)
    print(f"Number of clusters: {n_clusters}")

    # Cluster widths
    def calc_clusters_widths(
        X: np.array,
        centers: np.array,
        index: np.array,
        ra: float
    ) -> np.array:
        """
        Calculates sigma for each cluster
        """
        n_clusters = len(centers)
        n_features = X.shape[1]
        widths = np.zeros((n_clusters, n_features))

        if n_clusters > 1:
            center_distances = np.sqrt(np.sum((centers[:, np.newaxis] - centers[np.newaxis, :]) ** 2, axis=2))
        
            for i in range(n_clusters):

                other_dist = center_distances[i, :]
                other_dist[i] = np.inf
                min_distance = np.min(other_dist)
                widths[i, :] = min_distance / (2 * np.sqrt(2))

                min_allowed = ra/16
                widths[i, :] = np.maximum(widths[i, :], min_allowed)
        else:
            std_overall = np.std(X, axis=0)
            widths[0, :] = std_overall / 2

        return widths

    sigma_clusters = calc_clusters_widths(X_train_norm, cluster_centers, cluster_index, ra)

    # Membership matrix
    def calc_membership(
        X: np.array,
        centers: np.array,
        widths: np.array
    ) -> np.array:
        """
        Calculating membership matrix using gaussian functions
        """
        n_samples = X.shape[0]
        n_clusters = len(centers)
        U = np.zeros((n_samples, n_clusters))

        for i in range(n_clusters):
            diff = X - centers[i]
            # Mahalanobis
            normalized_dist = np.sum((diff / widths[i]) ** 2, axis=1)
            U[:, i] = np.exp(-.5 * normalized_dist)

        row_sums = U.sum(axis=1, keepdims=True)
        row_sums[row_sums == 0] = 1e-10
        U_normalized = U / row_sums

        return U_normalized

    U_train = calc_membership(X_train_norm, cluster_centers, sigma_clusters)
    print(f"Membership matrix shape: {U_train.shape}")

    # Defingin consequence parameter with MQ
    # y = Φ * b, where Φ = U_normalized
    # b = (ΦᵀΦ)⁻¹ Φᵀ y
    Phi_train = U_train
    b = np.linalg.pinv(Phi_train.T @ Phi_train) @ Phi_train.T @ y_train_norm

    def predict(
        X: np.array,
        centers: np.array,
        widths: np.array,
        b: np.array
    ) -> np.array:
        """
        Make predictions with the fuzzy model based on clustering method
        """
        U = calc_membership(X, centers, widths)
        y_pred_norm = U @ b
        return y_pred_norm

    # Predictions
    y_train_pred_norm = predict(X_train_norm, cluster_centers, sigma_clusters, b)
    y_test_pred_norm = predict(X_test_norm, cluster_centers, sigma_clusters, b)
    
    y_train_pred = y_train_pred_norm * y_std + y_mean
    y_test_pred = y_test_pred_norm * y_std + y_mean

    # Errors
    rmse_train = np.sqrt(np.mean((y_train - y_train_pred) ** 2))
    rmse_test = np.sqrt(np.mean((y_test - y_test_pred) ** 2))

    mae_train = np.mean(np.abs(y_train - y_train_pred))
    mae_test = np.mean(np.abs(y_test - y_test_pred))

    # R² score
    ss_total_train = np.sum((y_train - np.mean(y_train)) ** 2)
    ss_residual_train = np.sum((y_train - y_train_pred) ** 2)
    r2_train = 1 - (ss_residual_train / ss_total_train) if ss_total_train > 0 else 0

    ss_total_test = np.sum((y_test - np.mean(y_test)) ** 2)
    ss_residual_test = np.sum((y_test - y_test_pred) ** 2)
    r2_test = 1 - (ss_residual_test / ss_total_test) if ss_total_test > 0 else 0

    print(f"\nTraining RMSE: {rmse_train:.6f}")
    print(f"Testing RMSE: {rmse_test:.6f}")
    print(f"Training R²: {r2_train:.6f}")
    print(f"Testing R²: {r2_test:.6f}")

    print("\nCluster info")
    for i in range(n_clusters):
        print(f"Cluster {i+1}")
        print(f"Center: {cluster_centers[i]}")
        print(f"Widths: {sigma_clusters[i]}")
        print(f"Consequent b{i+1}: {b[i]:.6f}")

    return (cluster_centers, sigma_clusters, y_train_pred, y_test_pred, n_clusters, U_train, b, rmse_train, rmse_test, r2_train, r2_test)
```

```python
results = fuzzy_subtractive_clustering_method(
    X_train, y_train, X_test, y_test, ra=.5, rb=.75, epsilon=.01, max_clusters=4
)
```

```python
print(f"\n  Number of clusters found: {results[4]}")
print(f"  Training RMSE: {results[7]:.6f}")
print(f"  Testing RMSE:  {results[8]:.6f}")
print(f"  Training R²:   {results[9]:.6f}")
print(f"  Testing R²:    {results[10]:.6f}\n")

x_std = np.std(X_train, axis=0)
x_mean = np.mean(X_train, axis=0)
y_std = np.std(y_train, axis=0)
y_mean = np.mean(y_train, axis=0)

for i in range(results[4]):
    print(f"Region {i+1}")

    cluster_center = results[0][i] * x_std + x_mean
    sigma_cluster = results[1][i] * x_std
    b_params = results[6][i] * y_std + y_mean
    
    print(f"Typical Major Stress: {cluster_center[0]:.0f} +/- {sigma_cluster[0]:.0f} psi")
    print(f"Typical Minor Stress: {cluster_center[1]:.0f} +/- {sigma_cluster[1]:.0f} psi")
    print(f"Expected Deformation: {b_params*100:.2f}%")
```

C-MEANS CLUSTERING METHOD

C Means clustering is the ideia that every point belongs to every clusters, but with different memberships

```python
def fuzzy_cmeans_clustering_method(
    X_train: np.column_stack,
    y_train: np.array,
    X_test: np.column_stack,
    y_test: np.array,
    n_clusters: int = 5,
    m: float = 2.0,
    max_iter: int = 100,
    error: float = 1e-5
) -> np.array:
    """
    Implements a fuzzy model using Fuzzy C-Means Clustering

    1. Randomly initialization of the membership matrix
    2. Calculate cluster's centers
    3. Update the matrix
    4. Reppeat until the convergence

    Args
    -------
    X_train: array (N, 2)
        Train input
    y_train: array (N, )
        Train output
    X_test: array (M, 2)
        Test input
    y_test: array (M, )
        Test output
    n_clusters:
        Maximum number of clusters allowed
    m: Fuzzyness parameter
        m -> 1: more k-means-like
        m -> ∞: more fuzzy-like
    max_iter
        Maximum number of iteractions
    error
        Stop criteria (Convergence)

    Returns
    ---------
    clusters_centers: array (n_clusters, 2)
        Clusters centers found in the input space
    sigma_clusters: array (n_clusters, 2)
        Widths / Deviations of clusters in each dimension
    y_train_pred: array (N, )
        Training predictions
    y_test_pred: array (M, )
        Testing predictions
    n_clusters
        Number of clusters
    U: array (N, n_clusters)
        Membership matrix of the clusters
    """

    # Norm.
    x_mean = np.mean(X_train, axis=0)
    x_std = np.std(X_train, axis=0)
    X_train_norm = (X_train - x_mean) / x_std
    X_test_norm = (X_test - x_mean) / x_std
    
    y_mean = np.mean(y_train)
    y_std = np.std(y_train)
    y_train_norm = (y_train - y_mean) / y_std
    
    n_samples = X_train_norm.shape[0]
    n_features = X_train_norm.shape[1]
    
    # Random initialization of the matrix
    np.random.seed(42)
    U = np.random.rand(n_samples, n_clusters)
    U = U / np.sum(U, axis=1, keepdims=True)

    # loop fuzzy c-means
    for iteration in range(max_iter):
        U_old = U.copy()

        # Calc. cluster's centers
        U_m = U ** m
        cluster_centers = np.zeros((n_clusters, n_features))
        for i in range(n_clusters):
            numerator = np.sum(U_m[:, i:i+1] * X_train_norm, axis=0)
            denominator = np.sum(U_m[:, i])
            if denominator > 0:
                cluster_centers[i] = numerator / denominator

        # Calc. distances
        distances = np.zeros((n_samples, n_clusters))
        for i in range(n_clusters):
            diff = X_train_norm - cluster_centers[i]
            distances[:, i] = np.sum(diff ** 2, axis=1)
        distances[distances == 0] = 1e-10   # avoid zero division

        # Update matrix
        power = 1.0 / (m - 1)
        for i in range(n_clusters):
            for j in range(n_clusters):
                if i != j:
                    U[:, i] += (distances[:, i] / distances[:, j]) ** power
        U = 1.0 / U
        U = U / np.sum(U, axis=1, keepdims=True)

        # Verify convergence
        change = np.max(np.abs(U - U_old))
        if change < error:
            print(f"Converged - change = {change:.6f}")
            break
        if (iteration + 1) % 20 == 0:
            print(f"Iteration {iteration+1}: max change = {change:.6f}")

    # Calc. widths
    sigma_clusters = np.zeros((n_clusters, n_features))
    for i in range(n_clusters):
        weights = U[:, i] ** m
        weighted_mean = np.average(X_train_norm, axis=0, weights=weights)
        diff = X_train_norm - weighted_mean
        weighted_var = np.average(diff ** 2, axis=0, weights=weights)
        sigma_clusters[i] = np.sqrt(weighted_var)
        sigma_clusters[i] = np.maximum(sigma_clusters[i], 0.1)
    
    # Estimating b
    Phi_train = U
    b = np.linalg.pinv(Phi_train.T @ Phi_train) @ Phi_train.T @ y_train_norm

    def predict_fcm(
        X: np.array,
        centers: np.array,
        U_func: np.array,
        b: np.array,
        m: float
    ) -> np.array:
        """
        Make predictions with the fuzzy model
        """
        n_samples = X.shape[0]
        n_clusters = len(centers)
        
        distances = np.zeros((n_samples, n_clusters))
        for i in range(n_clusters):
            diff = X - centers[i]
            distances[:, i] = np.sum(diff ** 2, axis=1)
        distances[distances == 0] = 1e-10
        
        U_new = np.zeros((n_samples, n_clusters))
        for i in range(n_clusters):
            U_new[:, i] = 1.0
            for j in range(n_clusters):
                if i != j:
                    U_new[:, i] += (distances[:, i] / distances[:, j]) ** (1.0/(m-1))
        U_new = 1.0 / U_new
        U_new = U_new / np.sum(U_new, axis=1, keepdims=True)
        
        y_pred_norm = U_new @ b
        return y_pred_norm

    # Predictions
    y_train_pred_norm = predict_fcm(X_train_norm, cluster_centers, U, b, m)
    y_test_pred_norm = predict_fcm(X_test_norm, cluster_centers, U, b, m)
    
    y_train_pred = y_train_pred_norm * y_std + y_mean
    y_test_pred = y_test_pred_norm * y_std + y_mean
    
    # Metrics
    rmse_train = np.sqrt(np.mean((y_train - y_train_pred) ** 2))
    rmse_test = np.sqrt(np.mean((y_test - y_test_pred) ** 2))
    
    r2_train = 1 - np.sum((y_train - y_train_pred) ** 2) / np.sum((y_train - np.mean(y_train)) ** 2)
    r2_test = 1 - np.sum((y_test - y_test_pred) ** 2) / np.sum((y_test - np.mean(y_test)) ** 2)
    
    return (cluster_centers, sigma_clusters, y_train_pred, y_test_pred, n_clusters, U, b, rmse_train, rmse_test, r2_train, r2_test)
```

```python
results = fuzzy_cmeans_clustering_method(
    X_train, y_train, X_test, y_test, n_clusters=4, m=2.0, max_iter=100, error=1e-5
)
```

MODIFIED LEARNING FROM EXAMPLES

The MLFE adds momentum / addaptiveness to its membership functions and rules / parameters.

```python
def mlfe_fuzzy_model(
    X_train: np.column_stack,
    y_train: np.array,
    X_test: np.column_stack,
    y_test: np.array,
    n_sets: int = 3,
    initial_lr_c: float = .01,
    initial_lr_sigma: float = .001,
    initial_lr_b: float = .1,
    epochs: int = 2_000,
    tol: float = 1e-6,
    momentum: float = .9,
    lr_decay: float = .995,
    max_grad_norm: float = 5.0,
    patience: int = 100
) -> np.array:
    """
    Modified learning from examples

    1. Accelerate convergence, keeping a consistent direction
    2. Adjust learning rate with gradient behaviour
    3. Avoid explosive gradients (gradient clipping)
    4. Has it own early stopping
    5. Reduce exponentialy learning rate with the time

    Args
    --------
    X_train: array (N, 2)
        Train input
    y_train: array (N, )
        Train output
    X_test: array (M, 2)
        Test input
    y_test: array (M, )
        Test output
    n_sets : int (default=3)
        Number of fuzzy set for each variable
    initial_lr_* : float
        Initial learning rate for each parameter
    epochs : int (default=2000)
        Maximum number of iteractions
    tol : float (default=1e-6)
        Stop criteria (Convergence)
    momentum : float (default=0.9)
        Momentum factor
    lr_decay : float (default=0.995)
        Learning rate decay factor
    max_grad_norm : float (default=5.0)
        Maximumg clipping limit
    patience : int (default=100)
        Early stopping criteria

    Returns
    ---------
    dict_results: dict
        MLFE results in a dict
    """

    # Norm.
    x_mean = np.mean(X_train, axis=0)
    x_std = np.std(X_train, axis=0)
    X_train_norm = (X_train - x_mean) / x_std
    X_test_norm = (X_test - x_mean) / x_std
    
    y_mean = np.mean(y_train)
    y_std = np.std(y_train)
    y_train_norm = (y_train - y_mean) / y_std

    # Parameter initialization
    n_rules = n_sets * n_sets
    n_samples = X_train.shape[0]
    n_features = X_train.shape[1]
    
    c1 = np.linspace(np.min(X_train_norm[:, 0]), np.max(X_train_norm[:, 0]), n_sets)
    c2 = np.linspace(np.min(X_train_norm[:, 1]), np.max(X_train_norm[:, 1]), n_sets)
    
    sigma1 = np.ones(n_sets) * (np.max(X_train_norm[:, 0]) -  np.min(X_train_norm[:, 0])) / n_sets
    sigma2 = np.ones(n_sets) * (np.max(X_train_norm[:, 1]) -  np.min(X_train_norm[:, 1])) / n_sets
    
    b = np.zeros(n_rules)

    # Momentum
    v_c1 = np.zeros_like(c1)
    v_c2 = np.zeros_like(c2)
    v_sigma1 = np.zeros_like(sigma1)
    v_sigma2 = np.zeros_like(sigma2)
    v_b = np.zeros_like(b)

    # actual learning rates
    lr_c = initial_lr_c
    lr_sigma = initial_lr_sigma
    lr_b = initial_lr_b

    # Main loop - MLFE 

    losses = list()
    best_loss = np.inf
    best_epoch = 0
    no_improve_count = 0
    prev_grad_norm = None

    best_params = {
        'c1': c1.copy(), 'c2': c2.copy(),
        'sigma1': sigma1.copy(), 'sigma2': sigma2.copy(),
        'b': b.copy(),
        'loss': np.inf
    }

    for epoch in range(epochs):
        
        grad_c1 = np.zeros_like(c1)
        grad_c2 = np.zeros_like(c2)
        grad_sigma1 = np.zeros_like(sigma1)
        grad_sigma2 = np.zeros_like(sigma2)
        grad_b = np.zeros_like(b)
        
        total_loss = 0.0

        # batch
        for sample_idx in range(n_samples):
            x1 = X_train_norm[sample_idx, 0]
            x2 = X_train_norm[sample_idx, 1]
            y_true = y_train_norm[sample_idx]

            # forward pass -> membership and prediction
            activations = np.zeros(n_rules)
            phi = np.zeros(n_rules)

            rule_idx = 0
            for i in range(n_sets):
                for j in range(n_sets):
                    mu1 = np.exp(-0.5 * ((x1 - c1[i]) / sigma1[i]) ** 2)
                    mu2 = np.exp(-0.5 * ((x2 - c2[j]) / sigma2[j]) ** 2)
                    activations[rule_idx] = mu1 * mu2
                    rule_idx += 1

            sum_activations = np.sum(activations)
            if sum_activations > 0:
                phi = activations / sum_activations
            else:
                phi = np.ones(n_rules) / n_rules

            # Prediction
            y_pred = np.dot(phi, b)

            # Errors
            error = y_pred - y_true
            total_loss += .5 * (error ** 2)

            # beackward pass -> gradient
            grad_b += error * phi

            rule_idx = 0
            for i in range(n_sets):
                for j in range(n_sets):
                    mu1 = np.exp(-0.5 * ((x1 - c1[i]) / sigma1[i]) ** 2)
                    mu2 = np.exp(-0.5 * ((x2 - c2[j]) / sigma2[j]) ** 2)
                    phi_idx = phi[rule_idx]
                    b_idx = b[rule_idx]

                    common_term = error * (b_idx - y_pred) * (phi_idx / max(sum_activations, 1e-10))

                    if sigma1[i] != 0:
                        grad_c1[i] += common_term * mu2 * mu1 * (x1 - c1[i]) / (sigma1[i] ** 2)
                        grad_sigma1[i] += common_term * mu2 * mu1 * ((x1 - c1[i]) ** 2) / (sigma1[i] ** 3)
                    if sigma2[j] != 0:
                        grad_c2[j] += common_term * mu2 * mu1 * (x2 - c2[j]) / (sigma2[j] ** 2)
                        grad_sigma2[j] += common_term * mu2 * mu1 * ((x2 - c2[j]) ** 2) / (sigma2[j] ** 3)

                    rule_idx += 1
        
        grad_c1 /= n_samples
        grad_c2 /= n_samples
        grad_sigma1 /= n_samples
        grad_sigma2 /= n_samples
        grad_b /= n_samples

        # total gradient norm
        grad_norm = (
            np.linalg.norm(grad_c1) + np.linalg.norm(grad_c2) + np.linalg.norm(grad_sigma1) + np.linalg.norm(grad_sigma2) + np.linalg.norm(grad_b)
        )

        # Gradient clipping
        if grad_norm > max_grad_norm:
            
            scale = max_grad_norm / grad_norm
            
            grad_c1 *= scale
            grad_c2 *= scale
            grad_sigma1 *= scale
            grad_sigma2 *= scale
            grad_b *= scale
            
            grad_norm = max_grad_norm

        # learning rate adaptive
        if prev_grad_norm is not None:
            
            grad_ratio = grad_norm / max(prev_grad_norm, 1e-10)

            if grad_ratio > 2.0:
                lr_c *= .8
                lr_sigma *= .8
                lr_b *= .8

            elif grad_ratio < .5:
                lr_c *= .95
                lr_sigma *= .95
                lr_b *= .95

        prev_grad_norm = grad_norm

        # Momentum update
        v_c1 = momentum * v_c1 + lr_c * grad_c1
        v_c2 = momentum * v_c2 + lr_c * grad_c2
        v_sigma1 = momentum * v_sigma1 + lr_sigma * grad_sigma1
        v_sigma2 = momentum * v_sigma2 + lr_sigma * grad_sigma2
        v_b = momentum * v_b + lr_b * grad_b

        # Updating parameters with momentum
        c1 -= v_c1
        c2 -= v_c2
        sigma1 -= v_sigma1
        sigma2 -= v_sigma2
        b -= v_b

        # Parameter validation
        sigma1 = np.maximum(sigma1, 1e-3)
        sigma2 = np.maximum(sigma2, 1e-3)
        avg_loss = total_loss / n_samples
        losses.append(avg_loss)

        # Early stopping
        if avg_loss < best_loss:
            best_loss = avg_loss
            best_epoch = epoch
            best_params = {
                'c1': c1.copy(), 'c2': c2.copy(),
                'sigma1': sigma1.copy(), 'sigma2': sigma2.copy(),
                'b': b.copy(),
                'loss': avg_loss
            }
            no_improve_count = 0  # reset
        else:
            no_improve_count += 1

        # learning rate update
        lr_c *= lr_decay
        lr_sigma *= lr_decay
        lr_b *= lr_decay

        # Early stopping
        if epoch > 100 and (no_improve_count >= patience or \
                            (lr_c < tol * 100 and lr_sigma < tol * 100 and lr_b < tol * 100)):
            print("Early stop")
            if no_improve_count >= patience:
                print(f"No improvement for {patience} epochs")
            if lr_c < tol * 100 and lr_sigma < tol * 100 and lr_b < tol * 100:
                print(f"Learning rates < {tol*100:.1e}")
            break

        if (epoch+1) % 200 == 0:
            print(
                f"\nEpoch {epoch+1:4d}/{epochs}",
                f"Loss = {avg_loss:.6f}",
                f"LR = {lr_c:.2e}",
                f"Grad = {grad_norm:.3f}"
                )

    # Best parameters
    c1, c2 = best_params['c1'], best_params['c2']
    sigma1, sigma2 = best_params['sigma1'], best_params['sigma2']
    b = best_params['b']
    final_loss = best_params['loss']
    print(f"  \nFinal loss: {final_loss:.6f}")
    print(f"  Total epochs: {len(losses)}")

    def predict(
        X_norm: np.column_stack
    ) -> np.array:
        """
        Makes prediction with the fuzzy model
        """
        n_points = X_norm.shape[0]
        predictions_norm = np.zeros(n_points)
        
        for idx in range(n_points):
            x1 = X_norm[idx, 0]
            x2 = X_norm[idx, 1]
            
            activations = np.zeros(n_rules)
            rule_idx = 0
            
            for j in range(n_sets):
                for k in range(n_sets):
                    mu1 = np.exp(-0.5 * ((x1 - c1[j]) / sigma1[j]) ** 2)
                    mu2 = np.exp(-0.5 * ((x2 - c2[k]) / sigma2[k]) ** 2)
                    activations[rule_idx] = mu1 * mu2
                    rule_idx += 1
            
            sum_activations = np.sum(activations)
            if sum_activations > 0:
                phi = activations / sum_activations
            else:
                phi = np.ones(n_rules) / n_rules
            
            predictions_norm[idx] = np.dot(phi, b)
        
        return predictions_norm

    # Predictions
    y_train_pred_norm = predict(X_train_norm)
    y_test_pred_norm = predict(X_test_norm)

    y_train_pred = y_train_pred_norm * y_std + y_mean
    y_test_pred = y_test_pred_norm * y_std + y_mean

    # RMSE
    rmse_train = np.sqrt(np.mean((y_train - y_train_pred) ** 2))
    rmse_test = np.sqrt(np.mean((y_test - y_test_pred) ** 2))
    
    # R² Score
    ss_total_train = np.sum((y_train - np.mean(y_train)) ** 2)
    ss_residual_train = np.sum((y_train - y_train_pred) ** 2)
    r2_train = 1 - (ss_residual_train / ss_total_train) if ss_total_train > 0 else 0
    
    ss_total_test = np.sum((y_test - np.mean(y_test)) ** 2)
    ss_residual_test = np.sum((y_test - y_test_pred) ** 2)
    r2_test = 1 - (ss_residual_test / ss_total_test) if ss_total_test > 0 else 0
    
    # MAE
    mae_train = np.mean(np.abs(y_train - y_train_pred))
    mae_test = np.mean(np.abs(y_test - y_test_pred))
    
    print(f"  Training RMSE: {rmse_train:.6f}")
    print(f"  Testing RMSE:  {rmse_test:.6f}")
    print(f"  Training R²:   {r2_train:.6f}")
    print(f"  Testing R²:    {r2_test:.6f}")
    print(f"  Training MAE:  {mae_train:.6f}")
    print(f"  Testing MAE:   {mae_test:.6f}")

    results = {
        # Model
        'c1': c1, 'c2': c2,
        'sigma1': sigma1, 'sigma2': sigma2,
        'b': b,
        
        # Previsões
        'y_train_pred': y_train_pred,
        'y_test_pred': y_test_pred,
        
        # Metrics
        'rmse_train': rmse_train, 'rmse_test': rmse_test,
        'r2_train': r2_train, 'r2_test': r2_test,
        'mae_train': mae_train, 'mae_test': mae_test,
        
        # History
        'losses': losses,
        'best_epoch': best_epoch,
        'best_loss': best_loss,
        'n_rules': n_rules,
        
        # Statistics
        'x_mean': x_mean, 'x_std': x_std,
        'y_mean': y_mean, 'y_std': y_std,
        
        # Configuration
        'n_sets': n_sets,
        'method': 'MLFE',
        
        # MLFE parameters
        'momentum': momentum,
        'lr_decay': lr_decay,
        'patience': patience,
        'final_learning_rates': {'c': lr_c, 'sigma': lr_sigma, 'b': lr_b}
    }

    return results
```
